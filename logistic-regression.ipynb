Logistic regression is a classification model. Bounded by 0 to 1 it gives real value(Probability)

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
#Reading Data
Data= pd.read_csv('../input/breast-cancer-prediction-dataset/Breast_cancer_data.csv')
Data.head()
#Data.shape
```

```python
#Breaking data in to labels and Features
Labls= Data.diagnosis
Features= Data.loc[:, ('mean_radius', 'mean_texture', 'mean_perimeter', 'mean_area', 'mean_smoothness')]
Features.head()
Features.shape
```

```python
#Split data for training and testing
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split
Xtrain, Xtest, Ytrain, Ytest= train_test_split(Features, Labls, test_size=0.3, shuffle=True)
```

```python
#Importing Logistic regression from sklearn library
from sklearn.linear_model import LogisticRegression
Log_Reg=LogisticRegression()
```

```python
#Training Model
Log_Reg.fit(Xtrain, Ytrain)
```

```python
#Perdiction
Log_Reg.score(Xtrain, Ytrain)
```

```python
#Loop for taking average result (More accurate)
Array= 5*(np.arange(10))
Test_Score=[]
Train_Score=[]

for i in Array:
    Log_Reg= LogisticRegression(max_iter=i)
    Log_Reg.fit(Xtrain, Ytrain)
    Test_Score.append(Log_Reg.score(Xtest, Ytest))
    Train_Score.append(Log_Reg.score(Xtrain, Ytrain))
    
plt.xlabel("Iterations")
plt.ylabel("Accuracy")
plt.title("Logistic regression Accuracy graph")
plt.plot(Array, Train_Score, Label= 'Training Accuracy')
plt.plot(Array, Test_Score, Label= 'Testing Accuracy')
```

```python
#Normalizing data
Normalized_Data= (Data-Data.min())/(Data.max()-Data.min())
Normalized_Data.head()
```

```python
# Breaking data in to labels and features (Noramalized data)
Lbls= Normalized_Data.diagnosis
Ftrs= Normalized_Data.loc[:, ('mean_radius', 'mean_texture', 'mean_perimeter', 'mean_area', 'mean_smoothness')]
Ftrs.head()
```

```python
#Split data for training and testing
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split
xtrain, xtest, ytrain, ytest= train_test_split(Ftrs, Lbls, test_size=0.3, shuffle=True)
```

```python
#Model call
from sklearn.linear_model import LogisticRegression
Log_Reg=LogisticRegression()
Log_Reg.fit(xtrain, ytrain)
```

```python
#Perdiction
Log_Reg.score(xtrain, ytrain)
```

```python
#Loop for Normalized data
Array= 5*(np.arange(10))
test_Score=[]
train_Score=[]

for i in Array:
    Log_Reg= LogisticRegression(max_iter=i)
    Log_Reg.fit(xtrain, ytrain)
    test_Score.append(Log_Reg.score(xtest, ytest))
    train_Score.append(Log_Reg.score(xtrain, ytrain))
    
plt.xlabel("Iterations")
plt.ylabel("Accuracy")
plt.title("Logistic regression Accuracy graph")
plt.plot(Array, Train_Score, Label= 'Training Accuracy')
plt.plot(Array, Test_Score, Label= 'Testing Accuracy')
```

```python
# Plotting confusion matrix
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(Log_Reg, xtest, ytest)

# Classification report
from sklearn.metrics import classification_report
print(classification_report(Log_Reg.predict(xtest),ytest))
```

```python
#Importing Naive bayes model from sklearn
from sklearn.naive_bayes import GaussianNB
NB_Model=GaussianNB()
NB_Model.fit(xtrain, ytrain)
NB_Model.score(xtest, ytest)
```

```python
# Plotting confusion matrix
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(NB_Model, xtest, ytest)

# Classification report
from sklearn.metrics import classification_report
print(classification_report(NB_Model.predict(xtest),ytest))
```
